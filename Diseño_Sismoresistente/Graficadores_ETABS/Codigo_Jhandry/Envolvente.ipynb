```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
R = 7

# Especifica el nombre del archivo Excel
archivo_excel = 'Grafica12.xlsx'

# Especifica las celdas que deseas leer
celdas_a_leer_A = 'D4:D33'
celdas_a_leer_B = 'E4:E33'
celdas_a_leer_C = 'B4:B33'
celdas_a_leer_D = 'K4:K33'
celdas_a_leer_E = 'L4:L33'

# Utiliza la función read_excel para leer los datos de la fila A
Envolventexx = pd.read_excel(archivo_excel, usecols=celdas_a_leer_A, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila B
Envolventeyx = pd.read_excel(archivo_excel, usecols=celdas_a_leer_B, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila C
Alturas = pd.read_excel(archivo_excel, usecols=celdas_a_leer_C, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila D
Envolventexy = pd.read_excel(archivo_excel, usecols=celdas_a_leer_D, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila E
Envolventeyy = pd.read_excel(archivo_excel, usecols=celdas_a_leer_E, squeeze=True)

# Cálculo para inelástico
R = 0.75  # Se debe definir el valor de R
Envxx = Envolventexx * 0.75 * R
Envyx = Envolventeyx * 0.75 * R * -1
Envxy = Envolventexy * 0.75 * R
Envyy = Envolventeyy * 0.75 * R * -1

# Crear un DataFrame con los resultados
data = {
    'Alturas': Alturas,
    'Envxx': Envxx,
    'Envyx': Envyx,
    'Envxy': Envxy,
    'Envyy': Envyy
}
resultados = pd.DataFrame(data)

print(resultados)
```

```python


# Código previo para obtener Envxx, Envyx, Envxy, Envyy y Alturas

# Encontrar los máximos valores
max_Envxx = np.abs(np.max(Envxx))
max_Envyx = np.abs(np.min(Envyx))
max_Envxy = np.abs(np.max(Envxy))
max_Envyy = np.abs(np.min(Envyy))

# Textos para los máximos valores
texto_maximos = [
    f'X respecto Envx, deriva: {max_Envxx * 100:.4f}',
    f'Y respecto Envx, deriva: {max_Envyx * 100:.4f}',
    f'X respecto Envy, deriva: {max_Envxy * 100:.4f}',
    f'Y respecto Envy, deriva: {max_Envyy * 100:.4f}'
]

# Graficar
plt.figure()
colors = [[0.2, 0.4, 0.6], [0.8, 0.1, 0.3], [0.1, 0.7, 0.2], [0.9, 0.6, 0.1]]

plt.plot(Envxx, Alturas, 'o-', color=colors[0], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Envxx')
plt.plot(Envyx, Alturas, 'o-', color=colors[1], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Envyx')
plt.plot(Envxy, Alturas, 'o-', color=colors[2], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Envxy')
plt.plot(Envyy, Alturas, 'o-', color=colors[3], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Envyy')

plt.xlabel('Derivas [%]')
plt.ylabel('Altura [m]')
plt.title('Derivas Inelásticas, DS, TR475')
plt.grid(axis='y')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2, fontsize=8)

# Ubicación de los textos en el gráfico
pos_x = 0.05  # Cambia estos valores para ajustar la posición
pos_y = 0.95  # Cambia estos valores para ajustar la posición
plt.text(pos_x, pos_y, '\n'.join(texto_maximos), horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=8)

plt.tight_layout()
plt.show()
```
