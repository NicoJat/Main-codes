```python
import pandas as pd
R = 7

# Especifica el nombre del archivo Excel
archivo_excel = './Resultados_Excel/Cortante_piso.xlsx'

# Especifica las celdas que deseas leer
celdas_a_leer_A = 'F4:F63'
celdas_a_leer_B = 'G4:G63'
celdas_a_leer_C = 'B4:B63'
celdas_a_leer_D = 'M4:M63'
celdas_a_leer_E = 'N4:N63'

# Utiliza la función read_excel para leer los datos de la fila A
Cortxx = pd.read_excel(archivo_excel, usecols=celdas_a_leer_A, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila B
Cortyx = pd.read_excel(archivo_excel, usecols=celdas_a_leer_B, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila C
Alturas = pd.read_excel(archivo_excel, usecols=celdas_a_leer_C, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila D
Cortxy = pd.read_excel(archivo_excel, usecols=celdas_a_leer_D, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila E
Cortyy = pd.read_excel(archivo_excel, usecols=celdas_a_leer_E, squeeze=True)

# Crear un DataFrame con los resultados
data = {
    'Alturas': Alturas,
    'Cortxx': Cortxx,
    'Cortyx': Cortyx,
    'Cortxy': Cortxy,
    'Cortyy': Cortyy
}
resultados = pd.DataFrame(data)

print(resultados)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Código previo para obtener Cortxx, Cortyx, Cortxy, Cortyy y Alturas

# Graficar
plt.figure()
colors = [[0.2, 0.4, 0.6], [0.8, 0.1, 0.3], [0.1, 0.7, 0.2], [0.9, 0.6, 0.1]]

plt.plot(Cortxx, Alturas, 'o-', color=colors[0], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Cortxx')
plt.plot(Cortyx, Alturas, 'o-', color=colors[1], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Cortyx')
plt.plot(Cortxy, Alturas, 'o-', color=colors[2], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Cortxy')
plt.plot(Cortyy, Alturas, 'o-', color=colors[3], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Cortyy')

plt.xlabel('Cortante [T]')
plt.ylabel('Altura [m]')
plt.title('Cortante, DS, TR475')
plt.grid(axis='y')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2, fontsize=8)

# Encontrar los máximos valores
max_Cortxx = np.max(Cortxx)
max_Cortyx = np.abs(np.min(Cortyx))
max_Cortxy = np.max(Cortxy)
max_Cortyy = np.abs(np.min(Cortyy))

# Textos para los máximos valores
texto_maximos = [
    f'X respecto Envx, Cortante Máximo: {max_Cortxx:.2f}',
    f'Y respecto Envx, Cortante Máximo: {max_Cortyx:.2f}',
    f'X respecto Envy, Cortante Máximo: {max_Cortxy:.2f}',
    f'Y respecto Envy, Cortante Máximo: {max_Cortyy:.2f}'
]

# Ubicación de los textos en el gráfico
pos_x = 0.05  # Cambia estos valores para ajustar la posición
pos_y = 0.9  # Cambia estos valores para ajustar la posición
plt.text(pos_x, pos_y, '\n'.join(texto_maximos), horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=8)

plt.tight_layout()
plt.show()
```
