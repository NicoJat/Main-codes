```python
import pandas as pd
import matplotlib.pyplot as plt

# Especifica el nombre del archivo Excel y las celdas que deseas leer
archivo_excel = 'CMYCR2.xlsx'
celdas_a_leer = ['N4:N30', 'O4:O30', 'M4:M30', 'P4:P30', 'Q4:Q30']

# Leer los datos del archivo Excel
df = pd.read_excel(archivo_excel, sheet_name=None, usecols=celdas_a_leer)

# Renombrar columnas para mayor claridad
data = {
    'CMxx': df['Hoja1']['Column1'],
    'CMyx': df['Hoja1']['Column2'],
    'Alturas': df['Hoja1']['Column3'],
    'CMxy': df['Hoja1']['Column4'],
    'CMyy': df['Hoja1']['Column5']
}

# Crear el gráfico
plt.figure()
titulo = 'Coordenadas del Centro de Masa y Rigidez, DS, TR475'
plt.title(titulo)
plt.plot(data['CMyx'], data['Alturas'], 'b--', linewidth=1.5, label='CMxx')
plt.plot(data['CMyy'], data['Alturas'], 'r-.', linewidth=1.5, label='CMyy')

plt.grid(axis='x')
plt.xlabel('Coordenada X')
plt.ylabel('Altura [m]')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Especifica el nombre del archivo Excel y las celdas que deseas leer
archivo_excel = 'CMYCR2.xlsx'
celdas_a_leer = ['P4:P30', 'N4:N30', 'M4:M30', 'Q4:Q30']

# Leer los datos del archivo Excel
df = pd.read_excel(archivo_excel, sheet_name=None, usecols=celdas_a_leer)

# Renombrar columnas para mayor claridad
data = {
    'CMxx': df['Hoja1']['Column1'],
    'CMxy': df['Hoja1']['Column2'],
    'Alturas': df['Hoja1']['Column3'],
    'CMyx': df['Hoja1']['Column4']
}

# Crear el gráfico
plt.figure()
titulo = 'Coordenadas del Centro de Masa y Rigidez, DS, TR475'
plt.title(titulo)
plt.plot(data['CMxy'], data['Alturas'], 'g:', linewidth=1.5, label='CMxy')
plt.plot(data['CMxx'], data['Alturas'], 'm-', linewidth=1.5, label='CMxx')

plt.grid(axis='x')
plt.xlabel('Coordenada Y')
plt.ylabel('Altura [m]')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
