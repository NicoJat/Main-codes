```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
R = 7

# Especifica el nombre del archivo Excel
archivo_excel = 'Desplazamiento2.xlsx'

# Especifica las celdas que deseas leer
celdas_a_leer_A = 'D4:D33'
celdas_a_leer_B = 'E4:E33'
celdas_a_leer_C = 'B4:B33'
celdas_a_leer_D = 'K4:K33'
celdas_a_leer_E = 'L4:L33'

# Utiliza la función read_excel para leer los datos de la fila A
Desplazamientoenxdebidoax = pd.read_excel(archivo_excel, usecols=celdas_a_leer_A, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila B
Desplazamientoenydebidoax = pd.read_excel(archivo_excel, usecols=celdas_a_leer_B, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila C
Alturas = pd.read_excel(archivo_excel, usecols=celdas_a_leer_C, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila D
Desplazamientoenxdebidoay = pd.read_excel(archivo_excel, usecols=celdas_a_leer_D, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila E
Desplazamientoenydebidoay = pd.read_excel(archivo_excel, usecols=celdas_a_leer_E, squeeze=True)

# Cálculo para inelástico
R = 0.75  # Se debe definir el valor de R
Dxxine = Desplazamientoenxdebidoax * 0.75 * R
Dyxine = Desplazamientoenydebidoax * 0.75 * R * -1
Dxyine = Desplazamientoenxdebidoay * 0.75 * R
Dyyine = Desplazamientoenydebidoay * 0.75 * R * -1

# Crear un DataFrame con los resultados
data = {
    'Alturas': Alturas,
    'Dxxine': Dxxine,
    'Dyxine': Dyxine,
    'Dxyine': Dxyine,
    'Dyyine': Dyyine
}
resultados = pd.DataFrame(data)

print(resultados)
```

```python

# Código anterior para obtener Dxxine, Dyxine, Dxyine, Dyyine y Alturas

# Generar la figura
plt.figure()

# Definir colores
colors = [[0.2, 0.4, 0.6], [0.8, 0.1, 0.3], [0.1, 0.7, 0.2], [0.9, 0.6, 0.1]]

# Graficar cada conjunto de datos
plt.plot(Dxxine, Alturas, 'o-', color=colors[0], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Dxx')
plt.plot(Dyxine, Alturas, 'o-', color=colors[1], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Dyx')
plt.plot(Dxyine, Alturas, 'o-', color=colors[2], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Dxy')
plt.plot(Dyyine, Alturas, 'o-', color=colors[3], linewidth=1.5, markersize=3, markerfacecolor='auto', label='Dyy')

# Ajustes de la gráfica
plt.xlabel('Desplazamiento [m]')
plt.ylabel('Altura [m]')
plt.title('Desplazamiento Inelástico, DS, TR475')
plt.grid(axis='y')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2, fontsize=8)

# Calcula los valores máximos
max_Dxxine = np.max(Dxxine)
max_Dyxine = np.abs(np.min(Dyxine))
max_Dxyine = np.max(Dxyine)
max_Dyyine = np.abs(np.min(Dyyine))

# Texto para el cuadro de valores máximos
texto_maximos = [f'X respecto Envx, desp: {max_Dxxine:.4f} m',
                 f'Y respecto Envx, desp: {max_Dyxine:.4f} m',
                 f'X respecto Envy, desp: {max_Dxyine:.4f} m',
                 f'Y respecto Envy, desp: {max_Dyyine:.4f} m']

# Mostrar los valores máximos como leyenda
plt.text(0.5, -0.45, '\n'.join(texto_maximos), horizontalalignment='center', verticalalignment='center', fontsize=8, transform=plt.gca().transAxes)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
