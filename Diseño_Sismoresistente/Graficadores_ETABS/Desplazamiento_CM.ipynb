```python
import pandas as pd
R = 7

# Especifica el nombre del archivo Excel
archivo_excel = 'DesplazamientoCMCR2.xlsx'

# Especifica las celdas que deseas leer
celdas_a_leer_A = 'D4:D31'
celdas_a_leer_B = 'E4:E31'
celdas_a_leer_C = 'B4:B31'
celdas_a_leer_D = 'F4:F31'
celdas_a_leer_E = 'G4:G31'

# Utiliza la función read_excel para leer los datos de la fila A
DesplazamientoenxdebidoaxM = pd.read_excel(archivo_excel, usecols=celdas_a_leer_A, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila B
DesplazamientoenydebidoaxM = pd.read_excel(archivo_excel, usecols=celdas_a_leer_B, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila C
Alturas = pd.read_excel(archivo_excel, usecols=celdas_a_leer_C, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila D
DesplazamientoenxdebidoayM = pd.read_excel(archivo_excel, usecols=celdas_a_leer_D, squeeze=True)

# Utiliza la función read_excel para leer los datos de la fila E
DesplazamientoenydebidoayM = pd.read_excel(archivo_excel, usecols=celdas_a_leer_E, squeeze=True)

# Realizar los cálculos para inelástico
DCMXX = DesplazamientoenxdebidoaxM * 0.75 * R
DCMYX = DesplazamientoenydebidoaxM * 0.75 * R * -1
DCMXY = DesplazamientoenxdebidoayM * 0.75 * R
DCMYY = DesplazamientoenydebidoayM * 0.75 * R * -1

# Crear un DataFrame con los resultados
data = {
    'Alturas': Alturas,
    'DCMXX': DCMXX,
    'DCMYX': DCMYX,
    'DCMXY': DCMXY,
    'DCMYY': DCMYY
}
resultados = pd.DataFrame(data)

print(resultados)
```

```python
import matplotlib.pyplot as plt

# Calcular los valores máximos de cada conjunto de datos
max_DCMXX = max(DCMXX)
max_DCMYX = abs(min(DCMYX))
max_DCMXY = max(DCMXY)
max_DCMYY = abs(min(DCMYY))

# Textos para los máximos valores
texto_maximos = [
    f'X respecto Envx, desp: {max_DCMXX:.2f}',
    f'Y respecto Envx, desp: {max_DCMYX:.2f}',
    f'X respecto Envy, desp: {max_DCMXY:.2f}',
    f'Y respecto Envy, desp: {max_DCMYY:.2f}'
]

# Crear la figura y graficar los datos
plt.figure()
plt.title('Desplazamiento CM, DS, TR475')

plt.plot(DCMXX, Alturas, 'o-', color=[0.2, 0.4, 0.6], linewidth=1.5, markersize=3, markerfacecolor='auto', label='DCMXX')
plt.plot(DCMYX, Alturas, 'o-', color=[0.8, 0.1, 0.3], linewidth=1.5, markersize=3, markerfacecolor='auto', label='DCMYX')
plt.plot(DCMXY, Alturas, 'o-', color=[0.1, 0.7, 0.2], linewidth=1.5, markersize=3, markerfacecolor='auto', label='DCMXY')
plt.plot(DCMYY, Alturas, 'o-', color=[0.9, 0.6, 0.1], linewidth=1.5, markersize=3, markerfacecolor='auto', label='DCMYY')

plt.xlabel('Desplazamiento CM [m]')
plt.ylabel('Altura [m]')
plt.grid(True)
plt.legend(texto_maximos, loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=2, fontsize=8)

plt.show()
```
