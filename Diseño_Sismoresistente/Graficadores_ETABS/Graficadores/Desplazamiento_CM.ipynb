```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/Centros de Masa.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/Centros de Masa.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name='Story Response', header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name='Story Response', header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Desp_CM_XXmax = data_X['X-Dir']
Desp_CM_XXmin = data_X['X-Dir Min']
Desp_CM_XYmax = data_X['Y-Dir']
Desp_CM_XYmin = data_X['Y-Dir Min']
Desp_CM_YYmax = data_Y['Y-Dir']
Desp_CM_YYmin = data_Y['Y-Dir Min']
Desp_CM_YXmax = data_Y['X-Dir']
Desp_CM_YXmin = data_Y['X-Dir Min']

nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Desp_CM_XXmax': Desp_CM_XXmax[1:],
    'Desp_CM_XXmin': Desp_CM_XXmin[1:],
    'Desp_CM_XYmax': Desp_CM_XYmax[1:],
    'Desp_CM_XYmin': Desp_CM_XYmin[1:],
    'Desp_CM_YYmax': Desp_CM_YYmax[1:],
    'Desp_CM_YYmin': Desp_CM_YYmin[1:],
    'Desp_CM_YXmax': Desp_CM_YXmax[1:],
    'Desp_CM_YXmin': Desp_CM_YXmin[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Desp_CM_XXmax[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_XXmax', color='#9BD770')
plt.plot(Desp_CM_XXmin[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_XXmin', color='#66B032')
plt.plot(Desp_CM_XYmax[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_XYmax', color='#375F1B')
plt.plot(Desp_CM_XYmin[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_XYmin', color='#1B3409')
plt.xlabel('Desplazamientos CM en m')
plt.ylabel('Pisos')
plt.title('Desplazamientos del centro de masa')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
