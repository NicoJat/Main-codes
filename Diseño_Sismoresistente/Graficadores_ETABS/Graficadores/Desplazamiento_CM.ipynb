```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_Excel/des_CM_X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_Excel/des_CM_Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Desp_CM_X = data_X['X-Dir']
Desp_CM_Y = data_Y['Y-Dir']
nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Desp_CM_X': Desp_CM_X[1:],
    'Desp_CM_Y': Desp_CM_Y[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Desp_CM_X[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_SX_X')
plt.plot(Desp_CM_Y[1:], nodes_str, marker='o', linestyle='-', label='Desp_CM_SY_Y')
plt.xlabel('Desplazamientos CM en m')
plt.ylabel('Pisos')
plt.title('Desplazamientos del centro de masa')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
