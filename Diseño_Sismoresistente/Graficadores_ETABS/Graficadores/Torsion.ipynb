```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file_path = '../Resultados_Excel_P1/Ratios_Torsion.xlsx' 
sheet_name = 'Story Max Over Avg Disp'  

data = pd.read_excel(file_path, sheet_name=sheet_name, header=1)

X_EnvX = data[1:18]['Ratio'].iloc[::2].tolist()   #EL ULTIMO VALOR EN EL EXCEL -2
Y_EnvX = data[2:19]['Ratio'].iloc[::2].tolist()   #EL ULTIMO VALOR EN EL EXCEL -2
X_EnvY = data[37:54]['Ratio'].iloc[::2].tolist()  #EL ULTIMO VALOR EN EL EXCEL -2
Y_EnvY = data[38:55]['Ratio'].iloc[::2].tolist() #EL ULTIMO VALOR EN EL EXCEL -2

nodes = np.arange(9, 0, -1)
nodes_str = [str(node) for node in nodes]
nodes_str = nodes_str

data = {
    'Pisos': nodes_str,
    'X_EnvX': X_EnvX,
    'Y_EnvX': Y_EnvX,
    'X_EnvY': X_EnvY,
    'Y_EnvY': Y_EnvY
}

df = pd.DataFrame(data)
display(df)
```

```python

# Ancho de las barras
bar_width = 0.2

# Ajuste de posiciones para las barras en el eje x
pos_X_EnvX = [i - 1.5 * bar_width for i in range(len(nodes_str))]
pos_Y_EnvX = [i - 0.5 * bar_width for i in range(len(nodes_str))]
pos_X_EnvY = [i + 0.5 * bar_width for i in range(len(nodes_str))]
pos_Y_EnvY = [i + 1.5 * bar_width for i in range(len(nodes_str))]

plt.figure(figsize=(15, 6))

plt.bar(pos_X_EnvX, X_EnvX, width=bar_width, color='skyblue', label='X_EnvX')
plt.bar(pos_X_EnvY, X_EnvY, width=bar_width, color='orange', label='X_EnvY')
plt.bar(pos_Y_EnvX, Y_EnvX, width=bar_width, color='red', label='Y_EnvX')
plt.bar(pos_Y_EnvY, Y_EnvY, width=bar_width, color='green', label='Y_EnvY')

# Personalizar el gráfico
plt.xlabel('Piso')
plt.ylabel('n ratio')
plt.title('Factor de Amplificación de Torsion')
plt.legend()
plt.xticks(range(len(nodes_str)), nodes_str[::-1])  # Establecer las etiquetas en el eje x
plt.grid(axis='y')  # Mostrar líneas de la cuadrícula en el eje y
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout() 
plt.show()
```
