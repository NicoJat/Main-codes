```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/Shear X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/Shear Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name='Story Response', header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name='Story Response', header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Cortante_XXmax = data_X['X-Dir']
Cortante_XXmin = data_X['X-Dir Min']
Cortante_XYmax = data_X['Y-Dir']
Cortante_XYmin = data_X['Y-Dir Min']
Cortante_YYmax = data_Y['Y-Dir']
Cortante_YYmin = data_Y['Y-Dir Min']
Cortante_YXmax = data_Y['X-Dir']
Cortante_YXmin = data_Y['X-Dir Min']

nodes = [i // 2 if i % 2 == 0 else i // 2 + 1 for i in range(68, 0, -1)]
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Cortante_XXmax': Cortante_XXmax[1:],
    'Cortante_XXmin': Cortante_XXmin[1:],
    'Cortante_XYmax': Cortante_XYmax[1:],
    'Cortante_XYmin': Cortante_XYmin[1:],
    'Cortante_YYmax': Cortante_YYmax[1:],
    'Cortante_YYmin': Cortante_YYmin[1:],
    'Cortante_YXmax': Cortante_YXmax[1:],
    'Cortante_YXmin': Cortante_YXmin[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Cortante_XXmax[1:], nodes_str, marker='o', linestyle='-', label='Cortante_XXmax')
plt.plot(Cortante_XXmin[1:], nodes_str, marker='o', linestyle='-', label='Cortante_XXmin')
plt.plot(Cortante_XYmax[1:], nodes_str, marker='o', linestyle='-', label='Cortante_XYmax')
plt.plot(Cortante_XYmin[1:], nodes_str, marker='o', linestyle='-', label='Cortante_XYmin')
plt.xlabel('Cortante en T')
plt.ylabel('Pisos')
plt.title('Cortante')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
