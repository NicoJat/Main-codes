```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_Excel/Cortante_piso_X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_Excel/Cortante_piso_Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Cortante_X = data_X['X-Dir']
Cortante_Y = data_Y['Y-Dir']
nodes = [i // 2 if i % 2 == 0 else i // 2 + 1 for i in range(68, 0, -1)]
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Cortante_X': Cortante_X[1:],
    'Cortante_Y': Cortante_Y[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Cortante_X[1:], nodes_str, marker='o', linestyle='-', label='Cortante_X')
plt.plot(Cortante_Y[1:], nodes_str, marker='o', linestyle='-', label='Cortante_Y')
plt.xlabel('Cortante T')
plt.ylabel('Pisos')
plt.title('Cortante')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
