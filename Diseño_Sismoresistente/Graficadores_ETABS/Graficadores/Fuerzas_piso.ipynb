```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_Excel_P1/Fuerzas_X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_Excel_P1/Fuerzas_Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
FCX = data_X['X-Dir']
FCX = [float(item) for item in FCX[1:]]
FCY = data_Y['Y-Dir']
FCY = [float(item) for item in FCY[1:]]

Pisos = np.arange(10, 0, -1)
#Pisos = [str(node) for node in Pisos]
Pisos = [int(item) for item in Pisos]

data = {
    'Pisos': Pisos,
    'FCX': FCX,
    'FCY': FCY
}

df = pd.DataFrame(data)
display(df)

```

```python
FCX_array = np.array(FCX)
# Plotting the forces as arrows
plt.figure(figsize=(6, 8))
plt.quiver(FCX_array, Pisos, -FCX_array, np.zeros_like(Pisos), scale=1, scale_units='xy', angles='xy', color='blue')
plt.title('Distribuci√≥n de Fuerzas laterales con Sx direccion X')
plt.xlabel('Fuerza en T')
plt.ylabel('Pisos')
plt.grid('on')
plt.show()
```

```python
FCY_array = np.array(FCY)
# Plotting the forces as arrows
plt.figure(figsize=(6, 8))
plt.quiver(FCY_array, Pisos, -FCY_array, np.zeros_like(Pisos), scale=1, scale_units='xy', angles='xy', color='red')
plt.title('Distribuci√≥n de Fuerzas laterales con Sy direccion Y')
plt.xlabel('Fuerza en T')
plt.ylabel('Pisos')
plt.grid('on')
plt.show()
```
