```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
R = 5

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/centro de rigidez x.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/centro de rigidez y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name='Story Response', header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name='Story Response', header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Centro_Rigidez_XX = data_X['X-Dir']
Centro_Rigidez_XY = data_X['Y-Dir']
Centro_Rigidez_YY = data_Y['Y-Dir']
Centro_Rigidez_YX = data_Y['X-Dir']

nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]
nodes_str = nodes_str

data = {
    'Pisos': nodes_str,
    'Centro_Rigidez_XX': Centro_Rigidez_XX[1:],
    'Centro_Rigidez_XY': Centro_Rigidez_XY[1:],
    'Centro_Rigidez_YY': Centro_Rigidez_YY[1:],
    'Centro_Rigidez_YX': Centro_Rigidez_YX[1:]
}

df = pd.DataFrame(data)
display(df)
```

```python
# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Centro_Rigidez_XX[1:], nodes_str, marker='o', linestyle='-', label='Centro_Rigidez_XX')
plt.plot(Centro_Rigidez_XY[1:], nodes_str, marker='o', linestyle='-', label='Centro_Rigidez_XY')
plt.plot(Centro_Rigidez_YY[1:], nodes_str, marker='o', linestyle='-', label='Centro_Rigidez_YY')
plt.plot(Centro_Rigidez_YX[1:], nodes_str, marker='o', linestyle='-', label='Centro_Rigidez_YX')
plt.xlabel('Desp de Centros de Rigidez en m')
plt.ylabel('Pisos')
plt.title('Centros de Rigidez')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
