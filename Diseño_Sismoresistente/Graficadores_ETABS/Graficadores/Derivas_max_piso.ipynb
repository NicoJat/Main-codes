```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
R = 7

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_Excel/Derivas_Max_E_X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_Excel/Derivas_Max_E_Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Derivas_X = data_X['X-Dir']*R*0.75*100
Derivas_Y = data_Y['Y-Dir']*R*0.75*100
nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]
nodes_str = nodes_str

data = {
    'Pisos': nodes_str,
    'Derivas_X': Derivas_X[1:],
    'Derivas_Y': Derivas_Y[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Derivas_X[1:], nodes_str, marker='o', linestyle='-', label='Derivas_SX_X')
plt.plot(Derivas_Y[1:], nodes_str, marker='o', linestyle='-', label='Derivas_SY_Y')
plt.xlabel('Derivas en %')
plt.ylabel('Pisos')
plt.title('Derivas inelásticas')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
