```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
R = 5

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/Max Deriva X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/Max Deriva Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name='Story Response', header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name='Story Response', header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Derivas_XX = data_X['X-Dir']*R*0.75*100
Derivas_XY = data_X['Y-Dir']*R*0.75*100
Derivas_YY = data_Y['Y-Dir']*R*0.75*100
Derivas_YX = data_Y['X-Dir']*R*0.75*100

nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]
nodes_str = nodes_str

data = {
    'Pisos': nodes_str,
    'Derivas_XX': Derivas_XX[1:],
    'Derivas_XY': Derivas_XY[1:],
    'Derivas_YY': Derivas_YY[1:],
    'Derivas_YX': Derivas_YX[1:]
}

df = pd.DataFrame(data)
display(df)


```

```python
# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Derivas_XX[1:], nodes_str, marker='o', linestyle='-', label='Derivas_XX')
plt.plot(Derivas_XY[1:], nodes_str, marker='o', linestyle='-', label='Derivas_XY')
plt.plot(Derivas_YY[1:], nodes_str, marker='o', linestyle='-', label='Derivas_YY')
plt.plot(Derivas_YX[1:], nodes_str, marker='o', linestyle='-', label='Derivas_YX')
plt.xlabel('Derivas en %')
plt.ylabel('Pisos')
plt.title('Derivas inelásticas maximas')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
