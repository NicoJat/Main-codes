```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/Momento Volcante X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/Momento Volcante Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Mom_vol_XXmax = data_X['X-Dir']
Mom_vol_XXmin = data_X['X-Dir Min']
Mom_vol_XYmax = data_X['Y-Dir']
Mom_vol_XYmin = data_X['Y-Dir Min']
Mom_vol_YYmax = data_Y['Y-Dir']
Mom_vol_YYmin = data_Y['Y-Dir Min']
Mom_vol_YXmax = data_Y['X-Dir']
Mom_vol_YXmin = data_Y['X-Dir Min']

nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Mom_vol_XXmax': Mom_vol_XXmax[1:],
    'Mom_vol_XXmin': Mom_vol_XXmin[1:],
    'Mom_vol_XYmax': Mom_vol_XYmax[1:],
    'Mom_vol_XYmin': Mom_vol_XYmin[1:],
    'Mom_vol_YYmax': Mom_vol_YYmax[1:],
    'Mom_vol_YYmin': Mom_vol_YYmin[1:],
    'Mom_vol_YXmax': Mom_vol_YXmax[1:],
    'Mom_vol_YXmin': Mom_vol_YXmin[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Mom_vol_XXmax[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_XXmax')
plt.plot(Mom_vol_XXmin[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_XXmin')
plt.plot(Mom_vol_XYmax[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_XYmax')
plt.plot(Mom_vol_XYmin[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_XYmin')
plt.xlabel('Momentos de volcamiento en T*m')
plt.ylabel('Pisos')
plt.title('Momentos de volcamiento Sx')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
