```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_Excel/Momentos_volcamiento_X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_Excel/Momentos_volcamiento_Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Mom_vol_X = data_X['X-Dir']
Mom_vol_Y = data_Y['Y-Dir']
nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Mom_vol_X': Mom_vol_X[1:],
    'Mom_vol_Y': Mom_vol_Y[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Mom_vol_X[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_X')
plt.plot(Mom_vol_Y[1:], nodes_str, marker='o', linestyle='-', label='Mom_vol_Y')
plt.xlabel('Momentos de volcamiento en T*m')
plt.ylabel('Pisos')
plt.title('Momentos de volcamiento')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
