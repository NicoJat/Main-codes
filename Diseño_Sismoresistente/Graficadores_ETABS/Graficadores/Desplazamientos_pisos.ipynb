```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path_X = '../Resultados_ExcelF/excel sismos/Max Disp X.xlsx'  # Replace with your file path
file_path_Y = '../Resultados_ExcelF/excel sismos/Max Disp Y.xlsx'  # Replace with your file path
sheet_name = 'Story Response'  # Replace with your sheet name if different

# Load the data into a DataFrame
data_X = pd.read_excel(file_path_X, sheet_name=sheet_name, header=1)
data_Y = pd.read_excel(file_path_Y, sheet_name=sheet_name, header=1)

# Assuming your data has 'Node' and 'Displacement' columns
# Replace these column names with your actual column names
Max_Desp_XXmax = data_X['X-Dir']
Max_Desp_XXmin = data_X['X-Dir Min']
Max_Desp_XYmax = data_X['Y-Dir']
Max_Desp_XYmin = data_X['Y-Dir Min']
Max_Desp_YYmax = data_Y['Y-Dir']
Max_Desp_YYmin = data_Y['Y-Dir Min']
Max_Desp_YXmax = data_Y['X-Dir']
Max_Desp_YXmin = data_Y['X-Dir Min']

nodes = np.arange(34, 0, -1)
nodes_str = [str(node) for node in nodes]

data = {
    'Pisos': nodes_str,
    'Max_Desp_XXmax': Max_Desp_XXmax[1:],
    'Max_Desp_XXmin': Max_Desp_XXmin[1:],
    'Max_Desp_XYmax': Max_Desp_XYmax[1:],
    'Max_Desp_XYmin': Max_Desp_XYmin[1:],
    'Max_Desp_YYmax': Max_Desp_YYmax[1:],
    'Max_Desp_YYmin': Max_Desp_YYmin[1:],
    'Max_Desp_YXmax': Max_Desp_YXmax[1:],
    'Max_Desp_YXmin': Max_Desp_YXmin[1:]
}

df = pd.DataFrame(data)
display(df)

# Plotting
plt.figure(figsize=(8, 10))
plt.plot(Max_Desp_XXmax[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_XXmax')
plt.plot(Max_Desp_XYmax[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_XYmax')
plt.plot(Max_Desp_YYmax[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_YYmax')
plt.plot(Max_Desp_YXmax[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_YXmax')

#plt.plot(Max_Desp_XXmin[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_XXmin')
#plt.plot(Max_Desp_XYmin[1:], nodes_str, marker='o', linestyle='-', label='Max_Desp_XYmin')

plt.xlabel('Desplazamientos piso Sx en m')
plt.ylabel('Pisos')
plt.title('Desplazamientos del piso')
plt.legend()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()
```
