```python
import numpy as np
import Propseccion as prop


#Datos
E = 2040000 #Modulo de Elasticidad kg/cm2 
fy = 2530 #Esfuerzo de fluencia kg/cm2
fu = 4200 #Esfuerzo ultimo kg/cm2
G = 787220.92 #En kgf/cm2




perfil = prop.propseccioni(35, 40, 50, 3, 3, 2) # (bf1, bf2, hw, tf1, tf2, tw) (en cm)

Ag = perfil.area()
Ix = perfil.inerciaxi()
Iy = perfil.inerciayi()
J = perfil.inerciapi()
rx = perfil.rxi()
ry = perfil.ryi()
Cw = perfil.cwi()
#-------------------------------------------------------

print('Propiedades de la seccion:\n')
print(f'Area bruta: {Ag} cm2')
print(f'Inercia en X: {Ix} cm4')
print(f'Inercia en Y: {Iy} cm4')
print(f'Inercia polar: {J} cm4')
print(f'Radio de giro en X: {rx} cm')
print(f'Radio de giro en Y: {ry} cm')
print(f'Constante de alabeo: {Cw} cm6')
```

```python

#Calculo de propiedades de la seccion

Kx=1 
Ky=1 
Kz=1 
L=350  
In=np.minimum(Ix,Iy)
K=np.minimum(Kx,Ky)
r=np.minimum(rx,ry)
#------------------------------------


#Evaluacion de esbeltez

c=float(input('Especifique caso (Casos disponibles 1-9) segun Tabla B4-1a: '))

if c==1:
    b=float(input('Indicar b: '))   
    t=float(input('Indicar t: '))   #Espesor ala y alma
    lmb=b/t
    lmbr=0.56*(np.sqrt((E/fy)))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==2:
    b=float(input('Indicar b: '))
    tf=float(input('Indicar tf: '))  #Espesor del ala
    h=float(input('Indicar h: '))
    tw=float(input('Indicar tw: '))   #Espesor del alma
    kc=4/(np.sqrt(h/tw))
    if kc<0.35:
        kc=0.35
    elif kc>0.75:
        kc=0.75
    else:
        kc=4/(np.sqrt(h/tw))

    lmb=b/tf
    lmbr=0.64*(np.sqrt((kc*E/fy)))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==3:
    b=float(input('Indicar b: '))   
    t=float(input('Indicar t: '))   #Espesor ala y alma
    lmb=b/t
    lmbr=0.45*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==4:
    d=float(input('Indicar d: '))   
    t=float(input('Indicar t: '))   #Espesor
    lmb=d/t
    lmbr=0.75*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==5:
    h=float(input('Indicar h: '))   
    t=float(input('Indicar t: '))   #Espesor alma
    lmb=h/t
    lmbr=1.49*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==6:
    b=float(input('Indicar b: '))   
    t=float(input('Indicar t: '))   #Espesor ala y alma
    lmb=b/t
    lmbr=1.4*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==7:
    b=float(input('Indicar b: '))   
    t=float(input('Indicar t: '))   #Espesor ala y alma
    lmb=b/t
    lmbr=1.40*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==8:
    b=float(input('Indicar b: '))   
    t=float(input('Indicar t: '))  
    lmb=b/t
    lmbr=1.49*(np.sqrt(E/fy))
    if lmb>lmbr:
        scs='Slender'
    else:
        scs='non-slender'
elif c==9:
    D=float(input('Indicar D: '))   
    t=float(input('Indicar t: '))   
    lmb=D/t
    lmbr=0.11*(E/fy)
    if lmb>lmbr:
        scs='slender'
    else:
        scs='non-slender'




'''
Establecemos aqui las funciones E3, E4, E5, E6 y E7
correspondientes a cada capitulo de la AISC 360-16.
'''
def E3(K,L,r,Ag,E,fy,In):
    Fe = (((np.pi)**2)*E)/((K*L/r)**2)
    if ((K*L)/r)>(4.71*np.sqrt(E/fy)) or (fy/Fe)>2.25:
        e3Fcr = 0.877*Fe
    elif ((K*L)/r)<=(4.71*np.sqrt(E/fy)):
        e3Fcr = (0.658**(fy/Fe))*fy
    
    e3Pn=e3Fcr*Ag
    e3phiPn=0.9*e3Pn
    return e3Pn, e3phiPn, e3Fcr


def E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw):
    cso=input('Especifique caso de la seccion E4 (a,b o c):')
    if cso=='a':
        h0 = float(input('Especifique h0, la distancia entre centroides de las alas: '))
        Cw = (Iy*(h0**2))/4
        Lz = float(input('Especifique Lz de la seccion (cm): '))
        Kz = 1 #float(input('Especifique Kz de la seccion: '))
        Fe = ((((np.pi**2)*E*Cw)/((Kz*Lz)**2))+(G*J))/(Ix+Iy)
        if ((K*L)/r)>(4.71*np.sqrt(E/fy)) or (fy/Fe)>2.25:
            e4Fcr = 0.877*Fe
        elif ((K*L)/r)<=(4.71*np.sqrt(E/fy)):
            e4Fcr = (0.658**(fy/Fe))*fy
        
        e4Pn=e4Fcr*Ag
        e4phiPn=0.9*e4Pn
    elif cso=='b':
        x0= 18 #float(input('Especifique x0 (cm): '))
        y0= 12 #float(input('Especifique y0 (cm): '))
        r0=np.sqrt((x0**2)+(y0**2)+((Ix+Iy)/Ag))
        H=1-(((x0**2)+(y0**2))/(r0**2))
        Lz = float(input('Especifique Lz de la seccion (cm): '))
        Ly = float(input('Especifique Ly de la seccion (cm): '))
        Fey = (((np.pi)**2)*E)/((Ky*Ly/ry)**2)
        Fez = ((((np.pi**2)*E*Cw)/((Kz*Lz)**2))+(G*J))/(Ag*(r0**2))
        Fe=((Fey+Fez)/(2*H))*(1-(np.sqrt(1-((4*Fey*Fez*H)/((Fey+Fez)**2)))))
        if ((K*L)/r)>(4.71*np.sqrt(E/fy)) or (fy/Fe)>2.25:
            e4Fcr = 0.877*Fe
        elif ((K*L)/r)<=(4.71*np.sqrt(E/fy)):
            e4Fcr = (0.658**(fy/Fe))*fy
    
        e4Pn=e4Fcr*Ag
        e4phiPn=0.9*e4Pn
    elif cso=='c':
        x0= 18 #float(input('Especifique x0 (cm): '))
        y0= 12 #float(input('Especifique y0 (cm): '))
        r0=np.sqrt((x0**2)+(y0**2)+((Ix+Iy)/Ag))
        H=1-(((x0**2)+(y0**2))/(r0**2))
        Cw=float(input('Especifique Cw (cm6): '))
        Lx = float(input('Especifique Lx de la seccion (cm): '))
        Ly = float(input('Especifique Ly de la seccion (cm): '))
        Lz = float(input('Especifique Lz de la seccion (cm): '))
        Fex = (((np.pi)**2)*E)/((Kx*Lx/rx)**2)
        Fey = (((np.pi)**2)*E)/((Ky*Ly/ry)**2)
        Fez = ((((np.pi**2)*E*Cw)/((Kz*Lz)**2))+(G*J))/(Ag*(r0**2))
        a1=1-((x0**2)/(r0**2))-((y0**2)/(r0**2))
        b1=Fex+Fey+Fez+((Fey*(x0**2))/(r0**2))+((Fey*(y0**2))/(r0**2))
        c1=(Fey*Fez)+(Fex*Fez)
        d1=(Fex*Fey*Fez)
        coeff=[a1, -b1, c1, -d1]
        Fe=np.amin(np.roots(coeff))
        if ((K*L)/r)>(4.71*np.sqrt(E/fy)) or (fy/Fe)>2.25:
            e4Fcr = 0.877*Fe
        elif ((K*L)/r)<=(4.71*np.sqrt(E/fy)):
            e4Fcr = (0.658**(fy/Fe))*fy
    
        e4Pn=e4Fcr*Ag
        e4phiPn=0.9*e4Pn


    return e4Pn, e4phiPn, e4Fcr


def E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b):

    if c == 9:
        if (D/t) <= (0.11*(E/fy)):
            Ae=Ag
        elif (0.11*(E/fy))<(D/t) and (D/t)<(0.45*(E/fy)):
            Ae=(((0.038*E)/(fy*(D/t)))+(2/3))*Ag
        
        ePn=Fcr*Ae
        ephiPn=0.9*ePn
        print('Los siguientes valores consideran el Area efectiva debido a pandeo local')
        print(ePn)
        print(ephiPn)
    else: 
        if c==5 or c==7 or c==8 or c==9:
            c1=0.18
        elif c==6:
            c1=0.2
        else:
            c1= 0.22
        
        c2=(1-(np.sqrt(1-(4*c1))))/(2*c1)
        Fel = ((c2*(lmbr/lmb))**2)*fy

        if lmb<=lmbr*(np.sqrt(fy/Fcr)):
            be=b
        elif lmb>(lmbr*np.sqrt(fy/Fcr)):
            be=b*(1-(c1*np.sqrt(Fel/Fcr)))*(np.sqrt(Fel/Fcr))
        
        Ae=(b-be)*t
        ePn=Fcr*Ae
        ephiPn=0.9*ePn
        print('Los siguientes valores consideran el Area efectiva debido a pandeo local')
        print(ePn)
        print(ephiPn)
  
        
#E3(K,L,r,Ag,E,fy,In)
#E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In)

'''
En este apartado las secciones se encuentran
ordenadas secuencialmente segun la tabla E1.1.
especificar caso con2 la enumeracion corresponiente (1-9)
'''

#Calculo de resistencia flexo-compresion de la columna
cs=float(input('Especifique caso (1-9) en tabla E1.1: '))
if cs==1:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Pn=np.minimum(e3Pn,e4Pn)
        phiPn=np.minimum(e3phiPn,e4phiPn)
        print('Pn es: ',Pn)
        print('phiPn es: ',phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)
elif cs==2:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Pn=np.minimum(e3Pn,e4Pn)
        phiPn=np.minimum(e3phiPn,e4phiPn)
        print('Pn es: ',Pn)
        print('phiPn es: ',phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)
elif cs==3:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        print('Pn es: ',e3Pn)
        print('phiPn es: ',e3phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)
elif cs==4:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        print('Pn es: ',e3Pn)
        print('phiPn es: ',e3phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)
elif cs==5:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Pn=np.minimum(e3Pn,e4Pn)
        phiPn=np.minimum(e3phiPn,e4phiPn)
        print('Pn es: ',Pn)
        print('phiPn es: ',phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)
elif cs==6:
    if scs=='non-slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
        Pn=np.minimum(e3Pn,e4Pn)
        phiPn=np.minimum(e3phiPn,e4phiPn)
        print('Pn es: ',Pn)
        print('phiPn es: ',phiPn)
        #E6()
    elif scs=='slender':
        print('Descuide, seguimos trabajando en ello')
        #E6()
        #E7()
elif cs==7:
    if scs=='non-slender':
        print('Descuide, seguimos trabajando en ello')
        #E5()
    elif scs=='slender':
        print('Descuide, seguimos trabajando en ello')
        #E5()
elif cs==8:
    e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
    e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In,Cw)
    Pn=np.minimum(e3Pn,e4Pn)
    phiPn=np.minimum(e3phiPn,e4phiPn)
    print('Pn es: ',Pn)
    print('phiPn es: ',phiPn)
elif cs==9:
    if scs=='non-slender':
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In)
        print('Pn es: ',e4Pn)
        print('phiPn es: ',e4phiPn)
    elif scs=='slender':
        e3Pn, e3phiPn, e3Fcr=E3(K,L,r,Ag,E,fy,In)
        e4Pn, e4phiPn, e4Fcr=E4(Kx,Ky,Kz,G,J,rx,ry,K,L,r,Ag,E,fy,In)
        Fcr=np.minimum(e3Fcr,e4Fcr)
        E7(Fcr,fy,E,D,t,Ag,c,lmb,lmbr,b)




```
